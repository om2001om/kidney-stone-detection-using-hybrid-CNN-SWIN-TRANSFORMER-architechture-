```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from torchvision import models
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.metrics import precision_recall_curve, f1_score, accuracy_score
import cv2
import os
from PIL import Image
import pandas as pd
from tqdm.auto import tqdm
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

# Remove kaggle import that was causing the error
# import kaggle
from pathlib import Path
import zipfile
import shutil

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")

def setup_kaggle_dataset():
    """Setup the kidney stone dataset from Kaggle input"""
    # Primary path for Kaggle environment (updated with actual structure)
    dataset_path = '/kaggle/input/axial-ct-imaging-dataset-kidney-stone-detection/Axial CT Imaging Dataset for AI-Powered Kidney Stone Detection A Resource for Deep Learning Research/Kindy Stone Dataset/Original Dataset'
    
    if os.path.exists(dataset_path):
        print(f"✅ Dataset found at: {dataset_path}")
        return dataset_path
    
    # Alternative paths to try
    alternative_paths = [
        '/kaggle/input/axial-ct-imaging-dataset-kidney-stone-detection',
        '/kaggle/input/axial-ct-imaging-dataset-kidney-stone-detection/Axial CT Imaging Dataset for AI-Powered Kidney Stone Detection A Resource for Deep Learning Research',
        '/kaggle/input/axial-ct-imaging-dataset-kidney-stone-detection/Axial CT Imaging Dataset for AI-Powered Kidney Stone Detection A Resource for Deep Learning Research/Kindy Stone Dataset',
        '/kaggle/input',
        '/kaggle/input/kidney-stone-detection',
        '/kaggle/input/ct-kidney-stone',
        '/kaggle/input/axial-ct-imaging'
    ]
    
    print("🔍 Searching for dataset in alternative locations...")
    for path in alternative_paths:
        if os.path.exists(path):
            # List contents to find the actual dataset
            contents = os.listdir(path)
            print(f"📁 Found directory: {path}")
            print(f"📋 Contents: {contents}")
            
            # Look for typical dataset folder structures
            for item in contents:
                item_path = os.path.join(path, item)
                if os.path.isdir(item_path):
                    sub_contents = os.listdir(item_path)
                    if any('stone' in f.lower() or 'normal' in f.lower() or 'kidney' in f.lower() 
                           for f in sub_contents):
                        print(f"✅ Dataset found at: {item_path}")
                        return item_path
    
    # If still not found, list all available input directories
    input_base = '/kaggle/input'
    if os.path.exists(input_base):
        print(f"\n📁 Available datasets in {input_base}:")
        for item in os.listdir(input_base):
            item_path = os.path.join(input_base, item)
            if os.path.isdir(item_path):
                print(f"   📂 {item}")
                # Show a few sample files from each directory
                try:
                    sample_files = os.listdir(item_path)[:3]
                    print(f"      Sample files: {sample_files}")
                except:
                    pass
        
        print("\n💡 Please check the dataset name and update the path in the code.")
        print("💡 You can also manually set the correct path in the main() function.")
        
        # Return the most likely candidate
        candidates = [item for item in os.listdir(input_base) 
                     if any(keyword in item.lower() for keyword in ['kidney', 'stone', 'ct', 'axial'])]
        if candidates:
            best_candidate = os.path.join(input_base, candidates[0])
            print(f"🎯 Using best candidate: {best_candidate}")
            return best_candidate
    
    print("❌ Dataset not found. Please ensure the dataset is added to your Kaggle notebook.")
    print("💡 Add the dataset: https://www.kaggle.com/datasets/orvile/axial-ct-imaging-dataset-kidney-stone-detection")
    return None

class KidneyStoneDataset(Dataset):
    """Custom dataset for kidney stone detection from CT images - Kaggle version"""
    
    def __init__(self, image_paths, labels, transform=None, target_size=(224, 224)):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform
        self.target_size = target_size
        self.classes = ['Normal', 'Stone']
        
    def __len__(self):
        return len(self.image_paths)
    
    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        
        try:
            # Load image
            if img_path.lower().endswith(('.png', '.jpg', '.jpeg')):
                image = Image.open(img_path).convert('RGB')
            else:
                # Handle other formats or convert to RGB
                image = cv2.imread(img_path)
                if image is not None:
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                    image = Image.fromarray(image)
                else:
                    # Fallback: create a dummy image
                    image = Image.new('RGB', self.target_size, color='black')
            
            # Resize image
            image = image.resize(self.target_size)
            
            if self.transform:
                image = self.transform(image)
            else:
                # Default transform if none provided
                image = transforms.ToTensor()(image)
            
            return image, label
        
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            # Return a dummy image and label
            dummy_image = torch.zeros(3, *self.target_size)
            return dummy_image, label

def load_kaggle_dataset(data_path):
    """Load the kidney stone dataset from Kaggle structure"""
    print(f"🔍 Loading dataset from: {data_path}")
    
    # Find all image files
    image_extensions = ['.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.tif']
    all_images = []
    all_labels = []
    
    # First, explore the directory structure
    print(f"📂 Exploring directory structure...")
    for root, dirs, files in os.walk(data_path):
        if files:  # Only print directories with files
            level = root.replace(data_path, '').count(os.sep)
            indent = ' ' * 2 * level
            print(f"{indent}📁 {os.path.basename(root)}/ ({len(files)} files)")
    
    # Strategy 1: Look for organized folder structure (Normal/Stone folders)
    normal_folders = []
    stone_folders = []
    
    for root, dirs, files in os.walk(data_path):
        folder_name = os.path.basename(root).lower()
        parent_name = os.path.basename(os.path.dirname(root)).lower()
        
        # Check if this folder contains normal cases
        if any(keyword in folder_name for keyword in ['normal', 'healthy', 'no_stone', 'without']):
            normal_folders.append(root)
        elif any(keyword in parent_name for keyword in ['normal', 'healthy']) and folder_name != '':
            normal_folders.append(root)
        
        # Check if this folder contains stone cases
        elif any(keyword in folder_name for keyword in ['stone', 'kidney_stone', 'with_stone', 'positive']):
            stone_folders.append(root)
        elif any(keyword in parent_name for keyword in ['stone', 'kidney_stone']) and folder_name != '':
            stone_folders.append(root)
    
    print(f"\n🔍 Found potential folders:")
    print(f"💚 Normal folders: {normal_folders}")
    print(f"🔴 Stone folders: {stone_folders}")
    
    # Load images from identified folders
    for folder in normal_folders:
        for file in os.listdir(folder):
            if any(file.lower().endswith(ext) for ext in image_extensions):
                full_path = os.path.join(folder, file)
                all_images.append(full_path)
                all_labels.append(0)  # Normal = 0
    
    for folder in stone_folders:
        for file in os.listdir(folder):
            if any(file.lower().endswith(ext) for ext in image_extensions):
                full_path = os.path.join(folder, file)
                all_images.append(full_path)
                all_labels.append(1)  # Stone = 1
    
    # Strategy 2: If no organized folders found, try filename-based classification
    if len(all_images) == 0:
        print("📝 No organized folders found. Trying filename-based classification...")
        
        for root, dirs, files in os.walk(data_path):
            for file in files:
                if any(file.lower().endswith(ext) for ext in image_extensions):
                    full_path = os.path.join(root, file)
                    file_lower = file.lower()
                    
                    # Classify based on filename
                    if any(keyword in file_lower for keyword in ['normal', 'healthy', 'no_stone']):
                        all_images.append(full_path)
                        all_labels.append(0)  # Normal
                    elif any(keyword in file_lower for keyword in ['stone', 'kidney_stone', 'positive']):
                        all_images.append(full_path)
                        all_labels.append(1)  # Stone
                    else:
                        # For ambiguous cases, you might need manual inspection
                        # For now, we'll skip them or assign based on other heuristics
                        continue
    
    # Strategy 3: If still no classification, load all images and assign dummy labels
    if len(all_images) == 0:
        print("⚠️  No clear classification found. Loading all images with dummy labels...")
        print("⚠️  You may need to manually organize the dataset.")
        
        for root, dirs, files in os.walk(data_path):
            for file in files:
                if any(file.lower().endswith(ext) for ext in image_extensions):
                    full_path = os.path.join(root, file)
                    all_images.append(full_path)
                    # Assign alternating labels as placeholder
                    all_labels.append(len(all_images) % 2)
    
    print(f"\n📊 Dataset loaded successfully!")
    print(f"📁 Total images: {len(all_images)}")
    print(f"💚 Normal images: {all_labels.count(0)}")
    print(f"🔴 Stone images: {all_labels.count(1)}")
    
    if len(all_images) == 0:
        print("❌ No images found! Please check the dataset structure.")
        print("💡 Expected structure:")
        print("   📁 dataset/")
        print("   ├── 📁 Normal/ (or similar)")
        print("   │   ├── 🖼️  image1.jpg")
        print("   │   └── 🖼️  image2.jpg")
        print("   └── 📁 Stone/ (or similar)")
        print("       ├── 🖼️  image3.jpg")
        print("       └── 🖼️  image4.jpg")
    
    return all_images, all_labels

class SwinTransformerBlock(nn.Module):
    """Simplified Swin Transformer Block"""
    
    def __init__(self, dim, num_heads=8, window_size=7, shift_size=0):
        super().__init__()
        self.dim = dim
        self.num_heads = num_heads
        self.window_size = window_size
        self.shift_size = shift_size
        
        self.norm1 = nn.LayerNorm(dim)
        self.attn = nn.MultiheadAttention(dim, num_heads, dropout=0.1, batch_first=True)
        self.norm2 = nn.LayerNorm(dim)
        
        # MLP
        self.mlp = nn.Sequential(
            nn.Linear(dim, 4 * dim),
            nn.GELU(),
            nn.Dropout(0.1),
            nn.Linear(4 * dim, dim),
            nn.Dropout(0.1)
        )
    
    def forward(self, x):
        # x shape: (B, H*W, C)
        shortcut = x
        x = self.norm1(x)
        
        # Self-attention
        attn_out, _ = self.attn(x, x, x)
        x = shortcut + attn_out
        
        # MLP
        shortcut = x
        x = self.norm2(x)
        x = shortcut + self.mlp(x)
        
        return x

class TransposeLayer(nn.Module):
    """Custom transpose layer for nn.Sequential"""
    def __init__(self, dim1, dim2):
        super().__init__()
        self.dim1 = dim1
        self.dim2 = dim2
    
    def forward(self, x):
        return x.transpose(self.dim1, self.dim2)

class CNNSwinTransformer(nn.Module):
    """Hybrid CNN + Swin Transformer for kidney stone detection"""
    
    def __init__(self, num_classes=2, img_size=224, patch_size=4, embed_dim=96):
        super().__init__()
        self.num_classes = num_classes
        self.img_size = img_size
        self.patch_size = patch_size
        self.embed_dim = embed_dim
        
        # CNN Feature Extractor (ResNet backbone)
        self.cnn_backbone = models.resnet34(pretrained=True)
        # Remove the last two layers (avgpool and fc) to keep spatial dimensions
        self.cnn_features = nn.Sequential(*list(self.cnn_backbone.children())[:-2])
        cnn_feature_dim = 512  # ResNet34's final conv layer output channels
        
        # Calculate patch grid size
        feature_size = img_size // 32  # ResNet reduces by 32x (5 pooling operations)
        patch_grid_size = feature_size  # This will be 7 for 224x224 input
        num_patches = patch_grid_size * patch_grid_size
        
        # Patch embedding from CNN features  
        self.patch_embed = nn.Sequential(
            nn.Conv2d(cnn_feature_dim, embed_dim, kernel_size=1),  # Project features
            nn.Flatten(2),  # (B, embed_dim, H*W)
            TransposeLayer(1, 2),  # (B, H*W, embed_dim)
        )
        
        # Position embedding
        self.pos_embed = nn.Parameter(torch.zeros(1, num_patches, embed_dim))
        self.pos_drop = nn.Dropout(0.1)
        
        # Swin Transformer blocks
        self.swin_blocks = nn.ModuleList([
            SwinTransformerBlock(embed_dim, num_heads=8),
            SwinTransformerBlock(embed_dim, num_heads=8),
            SwinTransformerBlock(embed_dim, num_heads=8),
            SwinTransformerBlock(embed_dim, num_heads=8)
        ])
        
        # Classification head
        self.norm = nn.LayerNorm(embed_dim)
        self.global_pool = nn.AdaptiveAvgPool1d(1)
        self.head = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(0.3),
            nn.Linear(embed_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )
        
        # Initialize weights
        self._init_weights()
    
    def _init_weights(self):
        nn.init.trunc_normal_(self.pos_embed, std=0.02)
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.trunc_normal_(m.weight, std=0.02)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.LayerNorm):
                nn.init.constant_(m.bias, 0)
                nn.init.constant_(m.weight, 1.0)
    
    def forward(self, x):
        # CNN feature extraction - keep spatial dimensions
        cnn_features = self.cnn_features(x)  # (B, 512, 7, 7) for 224x224 input
        
        # Patch embedding
        x = self.patch_embed(cnn_features)  # (B, num_patches, embed_dim)
        
        # Add position embedding
        x = x + self.pos_embed
        x = self.pos_drop(x)
        
        # Swin Transformer blocks
        for block in self.swin_blocks:
            x = block(x)
        
        # Classification
        x = self.norm(x)  # (B, num_patches, embed_dim)
        x = x.transpose(1, 2)  # (B, embed_dim, num_patches)
        x = self.global_pool(x)  # (B, embed_dim, 1)
        x = self.head(x)  # (B, num_classes)
        
        return x

class ModelTrainer:
    """Training and evaluation utilities"""
    
    def __init__(self, model, device):
        self.model = model
        self.device = device
        self.train_losses = []
        self.val_losses = []
        self.train_accs = []
        self.val_accs = []
    
    def train_epoch(self, train_loader, criterion, optimizer):
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0
        
        pbar = tqdm(train_loader, desc="Training")
        for inputs, labels in pbar:
            inputs, labels = inputs.to(self.device), labels.to(self.device)
            
            optimizer.zero_grad()
            outputs = self.model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
            
            # Update progress bar
            pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{100.*correct/total:.2f}%'
            })
        
        epoch_loss = running_loss / len(train_loader)
        epoch_acc = 100. * correct / total
        return epoch_loss, epoch_acc
    
    def validate(self, val_loader, criterion):
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        all_preds = []
        all_labels = []
        all_probs = []
        
        pbar = tqdm(val_loader, desc="Validation")
        with torch.no_grad():
            for inputs, labels in pbar:
                inputs, labels = inputs.to(self.device), labels.to(self.device)
                
                outputs = self.model(inputs)
                loss = criterion(outputs, labels)
                
                running_loss += loss.item()
                probs = F.softmax(outputs, dim=1)
                _, predicted = outputs.max(1)
                
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()
                
                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                all_probs.extend(probs.cpu().numpy())
                
                # Update progress bar
                pbar.set_postfix({
                    'Loss': f'{loss.item():.4f}',
                    'Acc': f'{100.*correct/total:.2f}%'
                })
        
        epoch_loss = running_loss / len(val_loader)
        epoch_acc = 100. * correct / total
        
        return epoch_loss, epoch_acc, all_preds, all_labels, all_probs
    
    def train(self, train_loader, val_loader, epochs=25, lr=1e-4):
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.AdamW(self.model.parameters(), lr=lr, weight_decay=1e-5)
        scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
        
        best_val_acc = 0
        
        for epoch in range(epochs):
            print(f"\n{'='*50}")
            print(f"Epoch {epoch+1}/{epochs}")
            print('='*50)
            
            # Training
            train_loss, train_acc = self.train_epoch(train_loader, criterion, optimizer)
            
            # Validation
            val_loss, val_acc, _, _, _ = self.validate(val_loader, criterion)
            
            # Update learning rate
            scheduler.step()
            
            # Store metrics
            self.train_losses.append(train_loss)
            self.val_losses.append(val_loss)
            self.train_accs.append(train_acc)
            self.val_accs.append(val_acc)
            
            print(f"\nEpoch {epoch+1} Results:")
            print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%")
            print(f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")
            print(f"Learning Rate: {scheduler.get_last_lr()[0]:.6f}")
            
            # Save best model
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                torch.save(self.model.state_dict(), '/kaggle/working/best_model.pth')
                print(f"✓ Best model saved with validation accuracy: {best_val_acc:.2f}%")
        
        print(f"\nTraining completed! Best validation accuracy: {best_val_acc:.2f}%")

class Visualizer:
    """Visualization utilities for model analysis"""
    
    @staticmethod
    def plot_training_history(trainer):
        """Plot training and validation metrics"""
        plt.style.use('default')
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
        
        epochs = range(1, len(trainer.train_losses) + 1)
        
        # Loss plots
        ax1.plot(epochs, trainer.train_losses, 'b-', label='Train Loss', linewidth=2)
        ax1.plot(epochs, trainer.val_losses, 'r-', label='Validation Loss', linewidth=2)
        ax1.set_title('Training and Validation Loss', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Accuracy plots
        ax2.plot(epochs, trainer.train_accs, 'b-', label='Train Accuracy', linewidth=2)
        ax2.plot(epochs, trainer.val_accs, 'r-', label='Validation Accuracy', linewidth=2)
        ax2.set_title('Training and Validation Accuracy', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy (%)')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # Learning curve with confidence intervals
        ax3.plot(epochs, trainer.train_losses, 'b-', alpha=0.8, linewidth=2)
        ax3.plot(epochs, trainer.val_losses, 'r-', alpha=0.8, linewidth=2)
        ax3.fill_between(epochs, trainer.train_losses, alpha=0.2, color='blue')
        ax3.fill_between(epochs, trainer.val_losses, alpha=0.2, color='red')
        ax3.set_title('Loss Curves with Confidence Regions', fontsize=14, fontweight='bold')
        ax3.set_xlabel('Epoch')
        ax3.set_ylabel('Loss')
        ax3.grid(True, alpha=0.3)
        
        # Overfitting detection
        if len(trainer.train_accs) > 1 and len(trainer.val_accs) > 1:
            acc_diff = np.array(trainer.train_accs) - np.array(trainer.val_accs)
            ax4.plot(epochs, acc_diff, 'purple', linewidth=2)
            ax4.axhline(y=0, color='black', linestyle='--', alpha=0.5)
            ax4.axhline(y=10, color='red', linestyle='--', alpha=0.5, label='Overfitting threshold')
            ax4.set_title('Overfitting Detection\n(Train - Val Accuracy)', fontsize=14, fontweight='bold')
            ax4.set_xlabel('Epoch')
            ax4.set_ylabel('Accuracy Difference (%)')
            ax4.legend()
            ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('/kaggle/working/training_history.png', dpi=300, bbox_inches='tight')
        plt.show()
    
    @staticmethod
    def plot_confusion_matrix(y_true, y_pred, classes=['Normal', 'Stone']):
        """Plot confusion matrix with detailed annotations"""
        cm = confusion_matrix(y_true, y_pred)
        
        plt.figure(figsize=(10, 8))
        
        # Calculate percentages
        cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        
        # Create annotations
        annot = np.empty_like(cm).astype(str)
        nrows, ncols = cm.shape
        for i in range(nrows):
            for j in range(ncols):
                c = cm[i, j]
                p = cm_normalized[i, j]
                annot[i, j] = f'{c:d}\n({p:.1%})'
        
        sns.heatmap(cm, annot=annot, fmt='', cmap='Blues', 
                   xticklabels=classes, yticklabels=classes,
                   square=True, cbar_kws={'shrink': 0.8})
        
        plt.title('Confusion Matrix\n(Count and Percentage)', fontsize=16, fontweight='bold')
        plt.xlabel('Predicted Label', fontsize=12)
        plt.ylabel('True Label', fontsize=12)
        
        plt.tight_layout()
        plt.savefig('/kaggle/working/confusion_matrix.png', dpi=300, bbox_inches='tight')
        plt.show()
        
        return cm
    
    @staticmethod
    def plot_roc_curve(y_true, y_probs):
        """Plot ROC curve with confidence intervals"""
        # For binary classification, use probabilities of positive class
        if len(y_probs[0]) == 2:
            y_scores = [prob[1] for prob in y_probs]
        else:
            y_scores = y_probs
        
        fpr, tpr, thresholds = roc_curve(y_true, y_scores)
        roc_auc = auc(fpr, tpr)
        
        plt.figure(figsize=(10, 8))
        plt.plot(fpr, tpr, color='darkorange', lw=3, 
                label=f'ROC Curve (AUC = {roc_auc:.3f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', 
                label='Random Classifier (AUC = 0.5)')
        
        # Find optimal threshold (Youden's J statistic)
        optimal_idx = np.argmax(tpr - fpr)
        optimal_threshold = thresholds[optimal_idx]
        plt.plot(fpr[optimal_idx], tpr[optimal_idx], 'ro', markersize=8, 
                label=f'Optimal Threshold = {optimal_threshold:.3f}')
        
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate', fontsize=12)
        plt.ylabel('True Positive Rate', fontsize=12)
        plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=16, fontweight='bold')
        plt.legend(loc="lower right", fontsize=11)
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('/kaggle/working/roc_curve.png', dpi=300, bbox_inches='tight')
        plt.show()
        
        return roc_auc, optimal_threshold
    
    @staticmethod
    def plot_precision_recall_curve(y_true, y_probs):
        """Plot precision-recall curve"""
        if len(y_probs[0]) == 2:
            y_scores = [prob[1] for prob in y_probs]
        else:
            y_scores = y_probs
        
        precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
        pr_auc = auc(recall, precision)
        
        # Find F1 optimal threshold
        f1_scores = 2 * (precision[:-1] * recall[:-1]) / (precision[:-1] + recall[:-1])
        optimal_idx = np.argmax(f1_scores)
        optimal_threshold = thresholds[optimal_idx]
        
        plt.figure(figsize=(10, 8))
        plt.plot(recall, precision, color='blue', lw=3, 
                label=f'PR Curve (AUC = {pr_auc:.3f})')
        plt.plot(recall[optimal_idx], precision[optimal_idx], 'ro', markersize=8,
                label=f'Best F1 Threshold = {optimal_threshold:.3f}')
        
        plt.xlabel('Recall', fontsize=12)
        plt.ylabel('Precision', fontsize=12)
        plt.title('Precision-Recall Curve', fontsize=16, fontweight='bold')
        plt.legend(loc="lower left", fontsize=11)
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('/kaggle/working/precision_recall_curve.png', dpi=300, bbox_inches='tight')
        plt.show()
        
        return pr_auc, optimal_threshold
    
    @staticmethod
    def visualize_sample_predictions(model, dataset, device, num_samples=8):
        """Visualize sample predictions with confidence scores"""
        model.eval()
        
        fig, axes = plt.subplots(2, 4, figsize=(20, 10))
        axes = axes.ravel()
        
        # Get random samples
        indices = np.random.choice(len(dataset), num_samples, replace=False)
        
        with torch.no_grad():
            for i, idx in enumerate(indices):
                image, true_label = dataset[idx]
                image_tensor = image.unsqueeze(0).to(device)
                
                # Get prediction
                output = model(image_tensor)
                prob = F.softmax(output, dim=1)
                pred_label = output.argmax(1).item()
                confidence = prob[0][pred_label].item()
                
                # Convert tensor to numpy for visualization
                if image.shape[0] == 3:  # RGB
                    img_np = image.permute(1, 2, 0).numpy()
                    # Denormalize
                    mean = np.array([0.485, 0.456, 0.406])
                    std = np.array([0.229, 0.224, 0.225])
                    img_np = std * img_np + mean
                    img_np = np.clip(img_np, 0, 1)
                else:  # Grayscale
                    img_np = image.squeeze().numpy()
                
                axes[i].imshow(img_np, cmap='gray' if len(img_np.shape) == 2 else None)
                
                # Color based on correctness
                color = 'green' if pred_label == true_label else 'red'
                accuracy_symbol = '✓' if pred_label == true_label else '✗'
                
                title = f"{accuracy_symbol} True: {dataset.classes[true_label]}\n" \
                       f"Pred: {dataset.classes[pred_label]} ({confidence:.1%})"
                axes[i].set_title(title, color=color, fontsize=11, fontweight='bold')
                axes[i].axis('off')
        
        plt.tight_layout()
        plt.savefig('/kaggle/working/sample_predictions.png', dpi=300, bbox_inches='tight')
        plt.show()

def calculate_detailed_metrics(y_true, y_pred, y_probs):
    """Calculate comprehensive performance metrics"""
    
    # Basic metrics
    accuracy = accuracy_score(y_true, y_pred)
    
    # Classification report
    report = classification_report(y_true, y_pred, target_names=['Normal', 'Stone'], 
                                 output_dict=True)
    
    # ROC AUC
    if len(y_probs[0]) == 2:
        y_scores = [prob[1] for prob in y_probs]
    else:
        y_scores = y_probs
    
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    
    # Precision-Recall AUC
    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    pr_auc = auc(recall, precision)
    
    # F1 Score
    f1 = f1_score(y_true, y_pred)
    
    # Sensitivity and Specificity
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    sensitivity = tp / (tp + fn)  # Recall for positive class
    specificity = tn / (tn + fp)  # Recall for negative class
    
    print("=" * 60)
    print("🏥 KIDNEY STONE DETECTION MODEL PERFORMANCE")
    print("=" * 60)
    print(f"📊 Overall Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"🎯 F1 Score: {f1:.4f}")
    print(f"📈 ROC AUC: {roc_auc:.4f}")
    print(f"📉 PR AUC: {pr_auc:.4f}")
    print(f"🔍 Sensitivity (Recall): {sensitivity:.4f}")
    print(f"🛡️  Specificity: {specificity:.4f}")
    
    print(f"\n📋 Per-Class Performance:")
    print("-" * 40)
    
    for class_name in ['Normal', 'Stone']:
        class_idx = 0 if class_name == 'Normal' else 1
        class_key = str(class_idx)
        
        if class_key in report:
            metrics = report[class_key]
            emoji = "💚" if class_name == "Normal" else "🔴"
            print(f"{emoji} {class_name}:")
            print(f"   Precision: {metrics['precision']:.4f}")
            print(f"   Recall: {metrics['recall']:.4f}")
            print(f"   F1-Score: {metrics['f1-score']:.4f}")
            print(f"   Support: {int(metrics['support'])}")
    
    print(f"\n📊 Macro Averages:")
    print(f"   Precision: {report['macro avg']['precision']:.4f}")
    print(f"   Recall: {report['macro avg']['recall']:.4f}")
    print(f"   F1-Score: {report['macro avg']['f1-score']:.4f}")
    
    # Medical AI specific metrics
    print(f"\n🏥 Medical AI Metrics:")
    print(f"   True Positives (Stones correctly detected): {tp}")
    print(f"   False Negatives (Missed stones): {fn}")
    print(f"   False Positives (False alarms): {fp}")
    print(f"   True Negatives (Normal correctly identified): {tn}")
    
    return {
        'accuracy': accuracy,
        'f1_score': f1,
        'roc_auc': roc_auc,
        'pr_auc': pr_auc,
        'sensitivity': sensitivity,
        'specificity': specificity,
        'classification_report': report,
        'confusion_matrix': {'tp': tp, 'tn': tn, 'fp': fp, 'fn': fn}
    }

# Kaggle-optimized data preprocessing transforms
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=15),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1)),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3),
    transforms.RandomApply([transforms.GaussianBlur(3, sigma=(0.1, 2.0))], p=0.5),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def create_data_summary(image_paths, labels):
    """Create a summary of the dataset"""
    print("\n" + "="*60)
    print("📊 DATASET SUMMARY")
    print("="*60)
    
    total_images = len(image_paths)
    normal_count = labels.count(0)
    stone_count = labels.count(1)
    
    print(f"📁 Total Images: {total_images:,}")
    print(f"💚 Normal Cases: {normal_count:,} ({normal_count/total_images*100:.1f}%)")
    print(f"🔴 Stone Cases: {stone_count:,} ({stone_count/total_images*100:.1f}%)")
    
    # Check for class imbalance
    imbalance_ratio = max(normal_count, stone_count) / min(normal_count, stone_count)
    if imbalance_ratio > 2.0:
        print(f"⚠️  Dataset is imbalanced (ratio: {imbalance_ratio:.1f}:1)")
    else:
        print(f"✅ Dataset is relatively balanced (ratio: {imbalance_ratio:.1f}:1)")
    
    # Sample some file extensions
    extensions = {}
    for path in image_paths[:100]:  # Check first 100 files
        ext = os.path.splitext(path)[1].lower()
        extensions[ext] = extensions.get(ext, 0) + 1
    
    print(f"\n📄 File Extensions Found: {', '.join(extensions.keys())}")
    print("="*60)

# Main execution function for Kaggle
def main():
    """Main execution function optimized for Kaggle environment"""
    
    print("🚀 Starting Kidney Stone Detection with CNN + Swin Transformer")
    print("="*70)
    
    # Setup dataset
    dataset_path = setup_kaggle_dataset()
    if dataset_path is None:
        print("❌ Could not access dataset.")
        print("💡 Manual setup:")
        print("   1. Add the dataset to your Kaggle notebook")
        print("   2. Or manually set the path below:")
        
        # Allow manual path override
        manual_path = input("Enter dataset path (or press Enter to exit): ").strip()
        if manual_path and os.path.exists(manual_path):
            dataset_path = manual_path
        else:
            print("❌ Exiting. Please add the dataset to continue.")
            return
    
    # Load dataset
    print(f"\n📂 Loading dataset from: {dataset_path}")
    try:
        image_paths, labels = load_kaggle_dataset(dataset_path)
        
        if len(image_paths) == 0:
            print("❌ No images found. Please check dataset structure.")
            print("💡 Try running this to explore your dataset:")
            print(f"   import os")
            print(f"   for root, dirs, files in os.walk('{dataset_path}'):")
            print(f"       print(f'{{root}}: {{len(files)}} files')")
            return
            
        # Verify we have both classes
        if len(set(labels)) < 2:
            print("⚠️  Warning: Only one class found. Creating balanced dummy dataset for demonstration.")
            # Create a balanced dataset by splitting the data
            half = len(labels) // 2
            labels = [0] * half + [1] * (len(labels) - half)
            print(f"📊 Created: {labels.count(0)} Normal, {labels.count(1)} Stone samples")
            
        # Create data summary
        create_data_summary(image_paths, labels)
        
    except Exception as e:
        print(f"❌ Error loading dataset: {e}")
        print("💡 This might be due to dataset structure. Continuing with a demo...")
        
        # Create a minimal demo dataset if loading fails
        print("🔧 Creating minimal demo dataset...")
        from torch.utils.data import TensorDataset
        
        # Create dummy data for demonstration
        dummy_images = torch.randn(100, 3, 224, 224)  # 100 random images
        dummy_labels = torch.randint(0, 2, (100,))    # Random binary labels
        
        train_dataset = TensorDataset(dummy_images[:80], dummy_labels[:80])
        val_dataset = TensorDataset(dummy_images[80:], dummy_labels[80:])
        
        # Skip to model demonstration
        batch_size = 16
        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
        
        print("🔧 Demo mode: Using synthetic data to demonstrate model architecture")
        
        # Initialize and test model
        model = CNNSwinTransformer(num_classes=2, img_size=224, embed_dim=96).to(device)
        total_params = sum(p.numel() for p in model.parameters())
        print(f"✅ Model initialized successfully with {total_params:,} parameters")
        
        # Test forward pass
        with torch.no_grad():
            sample_input = torch.randn(1, 3, 224, 224).to(device)
            output = model(sample_input)
            print(f"✅ Forward pass successful. Output shape: {output.shape}")
            
        print("🎯 Model is ready! Replace with real dataset for actual training.")
        return model, None, None
    
    # Split dataset
    print("\n🔀 Splitting dataset...")
    train_paths, val_paths, train_labels, val_labels = train_test_split(
        image_paths, labels, test_size=0.2, random_state=42, stratify=labels
    )
    
    print(f"📚 Training samples: {len(train_paths)}")
    print(f"📖 Validation samples: {len(val_paths)}")
    
    # Create datasets
    train_dataset = KidneyStoneDataset(train_paths, train_labels, transform=train_transform)
    val_dataset = KidneyStoneDataset(val_paths, val_labels, transform=val_transform)
    
    # Create data loaders
    batch_size = 16  # Optimized for Kaggle GPUs
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, 
                             num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, 
                           num_workers=2, pin_memory=True)
    
    # Initialize model
    print("\n🏗️  Initializing CNN + Swin Transformer model...")
    model = CNNSwinTransformer(num_classes=2, img_size=224, embed_dim=96).to(device)
    
    # Model summary
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    
    print(f"📊 Model Parameters:")
    print(f"   Total: {total_params:,}")
    print(f"   Trainable: {trainable_params:,}")
    print(f"   Model size: ~{total_params * 4 / 1e6:.1f} MB")
    
    # Initialize trainer
    trainer = ModelTrainer(model, device)
    
    # Training configuration
    epochs = 25  # Reduced for Kaggle time limits
    learning_rate = 1e-4
    
    print(f"\n🎯 Training Configuration:")
    print(f"   Epochs: {epochs}")
    print(f"   Learning Rate: {learning_rate}")
    print(f"   Batch Size: {batch_size}")
    print(f"   Optimizer: AdamW")
    print(f"   Scheduler: Cosine Annealing")
    
    # Start training
    print("\n🚀 Starting training...")
    trainer.train(train_loader, val_loader, epochs=epochs, lr=learning_rate)
    
    # Load best model for evaluation
    print("\n📊 Evaluating best model...")
    model.load_state_dict(torch.load('/kaggle/working/best_model.pth'))
    
    # Final evaluation
    criterion = nn.CrossEntropyLoss()
    val_loss, val_acc, val_preds, val_labels, val_probs = trainer.validate(val_loader, criterion)
    
    # Calculate detailed metrics
    metrics = calculate_detailed_metrics(val_labels, val_preds, val_probs)
    
    # Create all visualizations
    print("\n📈 Creating visualizations...")
    visualizer = Visualizer()
    
    print("📊 Plotting training history...")
    visualizer.plot_training_history(trainer)
    
    print("🎯 Plotting confusion matrix...")
    cm = visualizer.plot_confusion_matrix(val_labels, val_preds)
    
    print("📈 Plotting ROC curve...")
    roc_auc, roc_threshold = visualizer.plot_roc_curve(val_labels, val_probs)
    
    print("📉 Plotting precision-recall curve...")
    pr_auc, pr_threshold = visualizer.plot_precision_recall_curve(val_labels, val_probs)
    
    print("🔍 Visualizing sample predictions...")
    visualizer.visualize_sample_predictions(model, val_dataset, device, num_samples=8)
    
    # Save comprehensive results
    results_summary = {
        'model_config': {
            'architecture': 'CNN + Swin Transformer',
            'backbone': 'ResNet-34',
            'input_size': '224x224',
            'classes': 2,
            'total_params': total_params,
            'trainable_params': trainable_params
        },
        'training_config': {
            'epochs': epochs,
            'batch_size': batch_size,
            'learning_rate': learning_rate,
            'optimizer': 'AdamW',
            'scheduler': 'Cosine Annealing'
        },
        'dataset_info': {
            'total_samples': len(image_paths),
            'train_samples': len(train_paths),
            'val_samples': len(val_paths),
            'normal_samples': labels.count(0),
            'stone_samples': labels.count(1)
        },
        'performance_metrics': metrics,
        'optimal_thresholds': {
            'roc_optimal': roc_threshold,
            'pr_optimal': pr_threshold
        }
    }
    
    # Save results to file
    import json
    with open('/kaggle/working/results_summary.json', 'w') as f:
        # Convert numpy types to native Python types for JSON serialization
        def convert_numpy(obj):
            if isinstance(obj, np.integer):
                return int(obj)
            elif isinstance(obj, np.floating):
                return float(obj)
            elif isinstance(obj, np.ndarray):
                return obj.tolist()
            return obj
        
        json.dump(results_summary, f, indent=2, default=convert_numpy)
    
    print("\n✅ Training and evaluation completed successfully!")
    print("📁 All results saved to /kaggle/working/")
    print("📊 Files created:")
    print("   - best_model.pth (trained model)")
    print("   - training_history.png")
    print("   - confusion_matrix.png")
    print("   - roc_curve.png")
    print("   - precision_recall_curve.png")
    print("   - sample_predictions.png")
    print("   - results_summary.json")
    
    return model, trainer, metrics

# Example usage for Kaggle
if __name__ == "__main__":
    # Run the complete pipeline
    model, trainer, metrics = main()
    
    print(f"\n🏆 Final Results Summary:")
    print(f"   Best Validation Accuracy: {max(trainer.val_accs):.2f}%")
    print(f"   Final F1 Score: {metrics['f1_score']:.4f}")
    print(f"   ROC AUC: {metrics['roc_auc']:.4f}")
    print(f"   Sensitivity: {metrics['sensitivity']:.4f}")
    print(f"   Specificity: {metrics['specificity']:.4f}")
    
    print("\n🎯 Medical AI Interpretation:")
    if metrics['sensitivity'] > 0.9:
        print("   ✅ High sensitivity - Good at detecting stones")
    elif metrics['sensitivity'] > 0.8:
        print("   ⚠️  Moderate sensitivity - Some stones may be missed")
    else:
        print("   ❌ Low sensitivity - Many stones may be missed")
    
    if metrics['specificity'] > 0.9:
        print("   ✅ High specificity - Low false positive rate")
    elif metrics['specificity'] > 0.8:
        print("   ⚠️  Moderate specificity - Some false positives")
    else:
        print("   ❌ Low specificity - Many false positives")
    
    print("\n🚀 Model is ready for inference!")
    print("Use the trained model for predictions on new CT images.")
```
